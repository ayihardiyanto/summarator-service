```python
import math
import pandas as pd
from math import log
```

```python
file = 'input.txt'
file = open(file , 'r')
text = file.read()

corpus = text.split("\n")
```

```python
def wordSetGenerator (corpus):
    listOfLines = []
    for line in corpus:
        print(len(line.split()))
        if line.strip():
            listOfLines.append(line.lower().split())
    return listOfLines;
            
generatedWords = wordSetGenerator(corpus);
```

```python
def wordSetUnifier(words):
    unifiedWords = '';
    for word in words:
        index = words.index(word)
        if(index == 0):
            unifiedWords = set(word);
        else:
            unifiedWords = unifiedWords.union(set(word))
    return unifiedWords;

wordSets = wordSetUnifier(generatedWords);
print(len(wordSets))
print(wordSets)
```

```python
def wordDictionaryGenerator(wordSets, generatedWords):
    dictionaries = []
    for word in generatedWords:
        dictionaries.append(dict.fromkeys(wordSets, 0))
    return dictionaries;

def initialValue(wordDict, generatedWords):
    for item in generatedWords:
        index = generatedWords.index(item)
        for word in item:
            wordDict[index][word] += 1

wordDicts = wordDictionaryGenerator(wordSets, generatedWords);
initialValue(wordDicts, generatedWords);
pd.DataFrame(wordDicts)
```

```python
def computeTF(wordDict, l):
    tf = {}
    sumNk = len(l)
    for word, count in wordDict.items():
        tf[word] = count/sumNk
    return tf


def computeIDF(strings_list):
    n = len(strings_list)
    idf = dict.fromkeys(strings_list[0].keys(), 0)
    for l in strings_list:
        for word, count in l.items():
            if count > 0:
                idf[word] += 1
    
    for word, v in idf.items():
        idf[word] = log(n / float(v))
    return idf


def computeTFIDF(tf, idf):
    tf_idf = dict.fromkeys(tf.keys(), 0)
    for word, v in tf.items():
        tf_idf[word] = v * idf[word]
    return tf_idf


def executeTFIDF():
    tfIdfs = []
    for item in generatedWords:
        index = generatedWords.index(item)
        tf = computeTF(wordDicts[index], item)
        idf = computeIDF(wordDicts)
        tfIdfs.append(computeTFIDF(tf, idf))
    return tfIdfs
    
```

```python
pd.DataFrame(executeTFIDF())
```
